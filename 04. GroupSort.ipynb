```python
import pandas as pd
```

```python
#https://raw.githubusercontent.com/GTD2101-ACD-AB-EJ23/03-pandas-VBatiz/main/data/winemag-data-130k-v2.csv
reviews = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-AB-EJ23/03-pandas-VBatiz/main/data/winemag-data-130k-v2.csv", index_col=0)
```

```python
pd.set_option("display.max_rows", 10)
```

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('country').country.count()
```

```python
reviews.groupby('points').price.max()
```

### Revisemos el nombre del primer vino evaluado de cada grupo de casas de vinos

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

### Mejor vino por país y provincia

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

### Índices múltiples

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
countries_reviewed = countries_reviewed.reset_index()
```

```python
countries_reviewed.head()
```

```python
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_values(by=['country', 'len'], ascending=False)
```
